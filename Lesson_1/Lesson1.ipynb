**Словари** <br>
Словарь - неупорядоченная структура данных, которая позволяет хранить пары «ключ — значение». 

```python
client_data = {
    'name': 'John',
    'surname': 'Doe',
    'age': 32
}

client_data
```

Чтобы достать нужный объект по ключу, к примеру ‘name’, нужно после имени словаря указать квадратные скобки и в них передать имя желаемого ключа.

```python
client_data['name']
```

Безопасный способ извлечения ключей, если их нет в списке, использовать метод **get()**. Тогда в случае отсутствия ключа, Python не будет выдавать ошибку -> на выходе ничего не получим, а все последующие ячейки кода будут запускаться без проблем и остановок.

```python
client_data.get('key')
```

Или можем получить все ключи, для этого вызываем метод keys(), возвращается список из ключей.

```python
client_data.keys()
```

По аналогии с ключами словаря, для получения всех значений можем вызвать метод values().

```python
client_data.values()
```

А  если хочется получить и ключи, и значения сразу, то тут используем items(), возвращается кортеж, где в первом элементе будет стоять ключ, а во втором элементе - значение.

```python
client_data.items()
```

Вот по этой структуре данных можно итерироваться, получая сразу ключи и значения за одну итерацию.

```python
for k, v in client_data.items():
    print(f'{k}: {v}')
```

Так же можем добавить новые сведения про John Doe в наши данные.

```python
client_data['height'] = 186
client_data
```

При этом если такой ключ уже был, то мы его перезапишем на новое значение, даже если мы этого не хотели.

```python
client_data['age'] = 14
client_data
```

Если боитесь перезаписать существующие ключи, то можете пользоваться методом **setdefault()**, тогда Python сначала посмотрит, а есть ли такой ключ в словаре и если он есть, то вернет его значение, а если такого ключа нет, то запишется на то, которое указано в методе. Видим, что у нас ничего не поменялось в ключе возраста, как там был возраст 32, так он и остался:

```python
client_data = {
    'name': 'John',
    'surname': 'Doe',
    'age': 32
}

client_data
```

```python
client_data.setdefault('age', 14)
```

Есть еще один интересный способ, как можно менять значения в словаре, через метод **update()**. Метод **update()** обновляет словарь элементами из другого объекта словаря или из итерируемых пар ключ-значение. Функция добавляет элемент(ы) в словарь, если ключ отсутствует в словаре. Если ключ находится в словаре, он обновляет ключ новым значением. <br>
И перезаписывает он не только существующие ключи, но и записывает новые, при этом здесь записывается вложенный словарь в ключ ‘body’. <br>

```python
client_data['body']['height']
```

**Аргумент args** <br>
*args — это сокращение от «arguments» (аргументы). <br>

Оператор * чаще всего ассоциируется у людей с операцией умножения, но в Python он имеет и другой смысл. Этот оператор позволяет «распаковывать» объекты, внутри которых хранятся некие элементы. <br>

Дело в том, что **«args»** — это всего лишь набор символов, которым принято обозначать аргументы. Самое главное тут — это оператор *. А то, что именно идёт после него, особой роли не играет. Благодаря использованию * мы создали список позиционных аргументов на основе того, что было передано функции при вызове.


Перепишем ранее созданную функцию суммирования, теперь для 3-х элементов:

```python
def calc_sum(a, b, c):
    result = a + b + c
    return result
```

Если хотим увеличить количество элементов, то снова нужно переписывать функцию:

```python
def calc_sum(a, b, c, d, e):
    result = a + b + c + d + e
    return result
```

Проще воспользоваться функцией ***args** и в дальнейшем вызывать ее с необходимым количеством элементов, без перезаписывания:

```python
def calc_sum(*args):
    result = sum(args)
    return result
```

```python
calc_sum(1, 2)
```

```python
calc_sum(10, 0, 5, 12, 2, 1, 59, 154)
```

Аргумент *args из себя представляет кортеж, который может быть любой длины, по которому можно итерироваться при желание, или же можно доставать объекты, или же можно просуммировать:

```python
def calc_sum(*args):
    print(args)
    print(type(args))
    result = sum(args)
    return result
```

```python
calc_sum(1, 2, 3)
```

Так же в args можно передавать элементы разных типов данных:

```python
def printer(*args):
    print(type(args))
    print(args)
    print()
    for i in args:
        print(i)
```

```python
printer(1, 'text', {'key': 'value'})
```

Но аккуратней со списками, потому что список - это один элемент:

```python
names = ['John', 'Nick', 'Bill']
printer(names)
```

Чтобы каждый элемент списка, считался самостоятельным аргументом нужно его распаковать, используя символ звездочки: 

```python
printer(*names)
```

**Генераторы**<br>

Допустим, у вас есть файл, который весит десяток гигабайт. Из него нужно выбрать и обработать строки, подходящие под какое-то условие. Что в такой ситуации делать? А можно обрабатывать такие объемы данных небольшими порциями, чтобы не вызывать переполнения памяти. В Python на этот случай есть специальный инструмент — генераторы.

> Генератор - это объект, который сразу при создании не вычисляет значения всех своих элементов.

Генератор хранит в памяти:

    1. последний вычисленный элемент
    2. правило перехода к следующему
    3. условие, при котором выполнение прерывается

Вычисление следующего значения происходит лишь при выполнении метода next(). Предыдущее значение при этом теряется. Этим генераторы отличаются от списков, ведь списки хранят в памяти все свои элементы, и удалить их можно только программно. Вычисления с помощью генераторов экономят память.

```python
def my_range(start, end):
    current = start
    while current < end:
        yield current
        current += 1
```

Оператор yield приостанавливает функцию и сохраняет локальное состояние, чтобы его можно было возобновить с того места, где оно было остановлено.<br>

Вызов функции не выполняет ее, а функция только возвращает объект-генератор:

```python
ranger = my_range(0, 2)
ranger
```

При первом вызове next() выполнение начинается с начала функции и продолжается до оператора yield, где возвращается значение справа от оператора:

```python
next(ranger)
```

Последующие вызовы next() продолжаются с оператора yield до конца функции, затем новый обход, который идет в начало функции и выполняется снова до оператора yeild:

```python
next(ranger)
```

Потом еще раз тоже самое, но если yield больше не вызывается (в нашем случае означает, что условие while не выполняется, потому что current >= 0), возникает исключение *StopIteration:*

```python
next(ranger)
```

По генератору можем итерироваться через цикл for. К примеру, создадим объект генератора, который реализовали выше и запишем в переменную gen_a. Теперь по нему можем пройтись в цикле, используя for, печатая каждый элемент на экран:

```python
gen_a = my_range(2, 5)
for i in gen_a:
    print(i)
```

А если захотим еще раз пройтись по этой же переменной, то ничего не вернется, это снова подтверждение тому, что генератор хранит только один элемент за раз, а не весь массив данных, как это делает список:

```python
for i in gen_a:
    print(i)
```

                            list comprehensions                             

Давайте вспомним про цикл for, если мы пользуемся им, чтобы из одного списка сделать новый список, то следуем этим пунктам:
1.	Создаем пустой список
2.	Итерируемся по объекту (или range)
3.	Добавляем каждый элемент в список


```python
squares = []
for i in range(5):
    squares.append(i * i)
squares
```

Или же можем пользоваться List comprehensions - это чудесный способ составления списков. Можем переписать цикл for всего в одну строку кода:

```python
squares = [i * i for i in range(5)]
squares
```

Вместо того, чтобы создавать пустой список и добавлять каждый элемент в конец, мы просто определяем список и его содержимое одновременно, следуя этому формату:<br>
            # new_list = [expression for member in iterable]        
где
-	expression - какое-то вычисление, вызов функции или любое другое допустимое выражение, которое возвращает значение.
-	member - объект или значение в списке или итерируемым объекте.
-	iterable - объект, по которому можно итерироваться (список, множество, последовательность, генератор)


                            Использование условий                           
Условные выражения позволяют отфильтровывать нежелательные значения. Снова сначала рассмотрим реализацию через цикл:


```python
even_nums = []
for i in range(10):
    if i % 2 == 0:
        even_nums.append(i)

even_nums
```

А теперь с помощью list comprehension:

```python
even_nums = [i for i in range(10) if i % 2 == 0]
even_nums
```

При этом условия могут быть сложнее, можем возвращать объект из итерируемого объекта или же заменять его на что-то, используя конструкцию if else:

```python
orig_prices = [125, -9, 103, 38, -5, 116]
prices = [i if i > 0 else 0 for i in orig_prices]
prices
```

Если условие хочется вынести отдельно, то можно реализовать дополнительную функцию для этого:

```python
def get_price(price):
    return price if price > 0 else 0


prices = [get_price(i) for i in orig_prices]
prices
```

                        set и dict comprehensions                           

Помимо list comprehension в Python еще возможно создавать множественные и словарные представления (**set comprehensions и dictionary comprehensions**).    

**set comprehension** почти точно такое же, как и **list comprehensions**. Разница лишь в том, что заданные значения обеспечивают, чтобы выходные данные не содержали дубликатов.     

Создать set comprehension можно, используя фигурные скобки вместо квадратных:

```python
numbers = ['black', 'red', 'green', 'green', 'red', 'green']
{i for i in numbers}
```

**Dict comprehensions** создаются аналогично, только еще добавляется определение ключа:

```python
squares_dict = {i: i * i for i in range(5)}
squares_dict
```

Можем замерить производительность подходов, **list comprehensions** *побеждают* обычные циклы:

```python
%%time

even_nums = []
for i in range(100_000_000):
    if i % 2 == 0:
        even_nums.append(i)
```

```python
%%time
even_nums = [i for i in range(100_000_000) if i % 2 == 0]
```
